```python
import numpy as np
```

```python
# Sample dataset (replace with your data)
data = [12, 15, 18, 20, 22, 25, 25, 27, 30, 35]
```

```python
# Calculate mean
mean = np.mean(data)
```

```python
# Calculate median
median = np.median(data)
```

```python
# Calculate standard deviation
std_dev = np.std(data)
```

```python
# NumPy doesn't have a built-in mode function, so we can use SciPy for that
from scipy import stats
```

```python
mode = stats.mode(data)
```

```python
print("Mean:", mean)
```

```python
print("Median:", median)
```

```python
print("Mode:", mode)
```

```python
print("Standard Deviation:", std_dev)
```
